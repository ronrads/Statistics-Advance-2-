# THEORY QUESTIONS

# Question 1: What is hypothesis testing in statistics?
"""Hypothesis testing is a statistical method to check if a claim about a population is likely true, based on data from a sample. It helps us decide whether to accept or reject an initial assumption (the null hypothesis) by evaluating the evidence."""

# Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
"""The null hypothesis is a statement of no effect or no difference, which is the default assumption. The alternative hypothesis is the statement a researcher wants to prove, suggesting a significant effect or difference exists."""

# Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
"""The significance level ($\alpha$) is the probability of rejecting the null hypothesis when it is actually true. It acts as a threshold for making a decision. If the p-value of a statistical test is less than or equal to the significance level, we reject the null hypothesis; otherwise, we fail to reject it."""

# Question 4: What are Type I and Type II errors? Give examples of each.
"""A Type I error, or false positive, occurs when you incorrectly reject a true null hypothesis, like concluding a drug works when it doesn't. A Type II error, or false negative, occurs when you incorrectly fail to reject a false null hypothesis, such as concluding a patient is healthy when they are actually sick. Both errors represent a wrong decision made during hypothesis testing."""

# Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
"""A Z-test and a T-test are both statistical tests used to compare means. The main difference lies in whether the population's standard deviation is known.

A Z-test is used when the population standard deviation is known or when the sample size is large (typically n > 30).

A T-test is used when the population standard deviation is unknown and the sample size is small (typically n < 30)."""

# Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5

num_samples = 10000
data = np.random.binomial(n, p, num_samples)

plt.figure(figsize=(10, 6))

plt.hist(data, bins=n + 1, rwidth=0.8, edgecolor='black', color='Green')

plt.title(f'Binomial Distribution Histogram (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')

plt.xticks(np.arange(n + 1))

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


```python
import numpy as np
from scipy import stats

sample_data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9])

mu_0 = 50.0
sigma = 1.0

z_statistic = (np.mean(sample_data) - mu_0) / (sigma / np.sqrt(len(sample_data)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
alpha = 0.05

print("Z-Test Results")
print("-" * 20)
print(f"Sample Mean: {np.mean(sample_data):.4f}")
print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")
print("-" * 20)

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
```

# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


np.random.seed(42)
simulated_data = np.random.normal(loc=100, scale=15, size=200)


confidence_interval = stats.norm.interval(0.95, loc=np.mean(simulated_data), scale=stats.sem(simulated_data))


print("Confidence Interval Results")
print("-" * 30)
print(f"Sample Mean: {np.mean(simulated_data):.2f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
print("-" * 30)


plt.figure(figsize=(10, 6))
plt.hist(simulated_data, bins=30, edgecolor='black', alpha=0.7, color='grey')
plt.axvline(x=np.mean(simulated_data), color='red', linestyle='--', label='Sample Mean')
plt.axvline(x=confidence_interval[0], color='green', linestyle=':', label='95% CI')
plt.axvline(x=confidence_interval[1], color='green', linestyle=':')
plt.title('Simulated Normal Distribution with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=10, size=500)


standardized_data = stats.zscore(sample_data)


print("Original Data Mean:", np.mean(sample_data).round(2))
print("Standardized Data Mean (should be ~0):", np.mean(standardized_data).round(2))
print("Standardized Data Std Dev (should be ~1):", np.std(standardized_data).round(2))


fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].hist(sample_data, bins=30, edgecolor='black', alpha=0.7, color='skyblue')
axes[0].set_title('Original Data')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

axes[1].hist(standardized_data, bins=30, edgecolor='black', alpha=0.7, color='lightgreen')
axes[1].set_title('Standardized Data (Z-scores)')
axes[1].set_xlabel('Z-Score')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
